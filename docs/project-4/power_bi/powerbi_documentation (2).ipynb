# Documentation: Creating Dynamic Dashboards with Power BI

## Overview

Power BI is a powerful business intelligence tool from Microsoft that enables data visualization,  
analysis, and reporting. This documentation provides a step-by-step guide on how to create and use  
dynamic dashboards like the visualizations provided.

## Step 1: Importing Data

- Click "Get Data" and select your data source (Excel, CSV, SQL, etc.).

- Load your dataset and ensure data types are correctly identified (e.g., Date fields, numerical  
  data) or if you want to transform the data, click on Transform data and write your  
  transformation.

![Image 1 - Transform Data](powerbi/Picture3.png)

After transforming click on close and apply  
Now , we can start creating visualisation

**Adding a new Column:**  
Go to the table view and click on the new column it will allow you to create a new column

![Image 2 - Add New Column](power_bi/powerbi/Picture4.png)

## Step 2: Creating Visualizations

### Example Visualizations

- **Pie Chart:**
  - Select Pie Chart from the visualization pane.
  - Drag "Day" to Legend and "First Time Visits" to Values.

![Image 3 - Pie Chart](powerbi/Picture5.png)

- **Bar Charts:**
  - Select Bar Chart from the visualization pane.
  - Drag "Day" to Axis and "First Time Visits" or "Returning Visits" to Values.

We can select the measure, we want to see on bar graph by clicking on y axis and  
another visual by new visual

![Image 4 - Bar Chart](powerbi/Picture6.png)

- **Line Charts (Trends Over Time):**
  - Select Line Chart from the visualization pane.
  - Drag "Year" or "Month" to Axis and measures such as "First Time Visits," "Returning  
    Visits," or "Unique Visits" to Values.

![Image 5 - Line Chart](powerbi/Picture8.png)

## Step 3: Applying Filters

- Use the "Filters" pane to dynamically interact with your dashboard.
- Apply filters by dragging fields into the "Filters on this page" or "Filters on all pages" sections.

![Image 6 - Filters](powerbi/Picture9.png)

## Step 4: Creating Dynamic Interactions

- Select any visual on your report canvas.
- Then go to the "Format" tab on the ribbon at the top (visible when a visual is selected).
- Click "Edit interactions". This enables options on all visuals to control how each one responds  
  to the selected visual (highlight, filter, or none).
- Clicking on a segment in one visual dynamically updates others based on related data.

![Image 7 - Edit Interactions](powerbi/Picture10.png)

## Step 5: Creating a Truly Dynamic Dashboard

To create a dynamic dashboard in Power BI:

- Use slicers for fields like "Date", "Category", or "Location" so users can filter data  
  dynamically.
- Create measures with DAX that change based on user inputs or slicer values (e.g., dynamic  
  titles, dynamic KPIs).
- Utilize Bookmarks and Buttons to switch between different views or toggle filters.
- Leverage drill-through pages for users to click on data points and view deeper details.

These techniques ensure your dashboard is not just static but interactive, customizable, and user-  
friendly.

## Step 6: Publishing Dashboards

- Click "Publish" to upload the dashboard to the Power BI Service, making it accessible online  
  and interactive for users.

## Conclusion

By following these steps and guidelines, you can effectively use Power BI to create dynamic,  
interactive dashboards, empowering insights and decision-making.
